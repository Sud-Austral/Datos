```python
import folium
import geopandas as gpd

fname = "https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/input/Cuarentenas/Cuarentenas-Geo.geojson"

states = gpd.read_file(fname)

#maploc = folium.Map(tiles="Stamen Toner")
maploc = folium.Map(location=[-33.6, -70.5], zoom_start=9.5)
#folium.GeoJson()

folium.GeoJson(
    #"https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/input/Cuarentenas/Cuarentenas-Geo.geojson",
    fname,
    name='geojson',
    style_function=lambda feature: {
        'fillColor': feature['properties']['Nombre'],
        'color' : feature['properties']['Nombre'],
        'weight' : 1,
        'fillOpacity' : 0.5,
        }
).add_to(maploc)
maploc
```

```python
import pandas as pd
salida = []
stringCoordenadas = str(states.geometry[1]).replace("POLYGON (("," ").replace("))","")
for i in stringCoordenadas.split(","):
    punto = i.split(" ")
    print(punto[1]+punto[2])
    aux = {
            "Latitud" : punto[2].replace(".",","),
            "Longitud" : punto[1].replace(".",",")
            }
    salida.append(aux.copy())
pd.DataFrame(salida).to_csv("Coordenadas.csv", index=False)
```
