```python
import pandas as pd

data_paises = pd.read_csv("../InformacionGeneral/Internacional/Detalle_paises.csv")
del data_paises["WORLD_SN"]
del data_paises["ECDC_SN"]
del data_paises["OURWOR_SN"]
del data_paises["JHONH_SN"]
del data_paises["Early_SN"]

data_woldmeter = pd.read_csv("../worldometers.info/worldometers.csv")
columna_woldmeter = data_woldmeter.columns[1:]

for i in columna_woldmeter:
    data_woldmeter[i+"Worldmeter"] = data_woldmeter[i]
    del data_woldmeter[i]
    
merged_left = pd.merge(left=data_paises, right=data_woldmeter, how='left', left_on='WORLDM_Country,Other', right_on='Country,Other')

#data_europa = pd.read_csv("../ecdc.europa/ecdc_europa.csv")
#data_europa
data_ourworld_con = pd.read_csv("../ourworldindata.org/total-cases-covid-19.csv")
data_ourworld_dea = pd.read_csv("../ourworldindata.org/total-deaths-covid-19.csv")
#data_ourworld["Entity"].unique()
#data_ourworld
paises = []
confirmados = []
muertes = []
for i in data_ourworld_con["Entity"].unique():
    maximoConfirmado = max(data_ourworld_con[data_ourworld_con["Entity"] == i]["Total confirmed cases of COVID-19 (cases)"])
    maximoMuerte = max(data_ourworld_dea[data_ourworld_dea["Entity"] == i]["Total confirmed deaths due to COVID-19 (deaths)"])
    #print(i + " " + str(maximo))
    paises.append(i)
    confirmados.append(maximoConfirmado)
    muertes.append(maximoMuerte)

data_ourworld = pd.DataFrame({
    "paises": paises,
    "confirmadoOURWORLD":confirmados,
    "muertosOURWORLD": muertes    
}
)
merged_left = pd.merge(left=merged_left, right=data_ourworld, how='left', left_on='OURWORLD_Entity', right_on='paises')
del merged_left["paises"] 
merged_left
```
